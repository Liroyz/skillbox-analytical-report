```python
### YOUR CODE HERE ###

# Считываем данные из CSV-файлов
students = pd.read_csv('Data/students.csv', parse_dates=['birthday'], dayfirst=False)
#Создаем оотдельный датафрейм для расчета среднего возраста студентов курсов
averege_age_students = pd.read_csv('Data/students.csv', parse_dates=['birthday'], dayfirst=False)

courses = pd.read_csv('Data/courses.csv', index_col=[0])
course_contents = pd.read_csv('Data/course_contents.csv')
progresses = pd.read_csv('Data/progresses.csv')
progress_phases = pd.read_csv('Data/progress_phases.csv')

# Переименуем колонку из 'id' в 'progress_id' в  датафрейме progresses для более удобного объединения 
progresses.rename(columns={ progresses.columns[0]: 'progress_id' }, inplace = True)

# Соединяем progress_phases с progresses
merged_progress = pd.merge(progress_phases, progresses, how='left', left_on='progress_id', right_on='progress_id')

# Переименуем колонку из 'id' в 'course_id' в  датафрейме courses для более удобного объединения 
courses.rename(columns={ courses.columns[0]: 'course_id' }, inplace = True)

# Присоединяем courses
merged_1 = pd.merge(merged_progress, courses, how='left', left_on='course_id', right_on='course_id')
merged_all = pd.merge(merged_1, course_contents, how='left')
merged_all
```

```python
# Создаем датафрейм взяв уникальные значения колонки названия курсов 'title' из основного датафрейма
courses_count = pd.DataFrame(merged_all['title'].unique())
courses_count
```

```python
# Создаем датафрейм сгруппированный по колонкам 'field' и 'title' вычисляя колличество 
# уникальных значений module_number
module_count = pd.DataFrame(merged_all.groupby(['field','title'])['module_number']\
                            .agg('nunique').reset_index())
module_count
```

### 1.1.3 Расчет количества уроков в каждом модуле на каждом курсе.

```python
# Создаем датафрейм сгруппированный по колонкам 'field', 'title', 'module_number' вычисляя количество 
# уникальных значений в колонке 'lesson_number'
lesson_count = pd.DataFrame(merged_all.groupby(['field', 'title', 'module_number'])['lesson_number']\
                            .agg('nunique').reset_index())
lesson_count
```

### 1.1.5 Расчет количества студентов на каждом курсе.

```python
# Создаем датафрейм сгруппированный по колонке области деятельности 'field' и колонке названия курсов 'title'
# агреггируя по колонке идентификаторов студентов 'student_id' с параметром 'nunique' 
# для вычисления кол-ва студентов
student_count = pd.DataFrame(merged_all.groupby(['field','title'])['student_id']\
                            .agg('nunique').reset_index())

# Сортируем по по колонке 'student_id' по убыванию
student_count.sort_values('student_id', ascending=False)
```

### 1.1.6 Расчет минимального, максимального, среднего, медианного возраста студентов.

```python
# Выкидываем NaN-ы из колонки 'birthday' и форматируем даты
students_birthday = students['birthday'].dropna().apply(lambda x: dt.date.fromisoformat(x))

# Отфильтровываем некорректные даты рождения и корректные и вычисляем количество полных лет студентов
students_birthday = students_birthday[students_birthday>dt.date(1941, 1, 1)][students_birthday<dt.date(2010, 1, 1)].apply((lambda x: (dt.date.today() - x).days/365.25))
students_birthday
```

### 1.1.7 Расчет минимального, максимального, среднего, медианного возраста студентов на каждом курсе.

```python
# Для начала из общего смерджинного датафрейма возьмем колонку названия курсов 'title' 
# и поле идентификаторов студентов
course_title = pd.DataFrame(merged_all[['title', 'student_id']])

# В копии датафрейма из файла Students.csv переименуем колонку 'id' в 'student_id'
averege_age_students.rename(columns={ averege_age_students.columns[1]: 'student_id' }, inplace = True)

# Мерджим два этих датафрейма в один, взяв из averege_age_students только колонки 'student_id' и 'birthday'
av_age = pd.merge(course_title, averege_age_students[['student_id', 'birthday']], 
                  how='left', left_on='student_id', right_on='student_id')

# Выкидываем NaN-ы из колонки 'birthday' и форматируем даты 
av_age['birthday'] = av_age['birthday'].dropna().apply(lambda x: dt.date.fromisoformat(x))

# Отфильтровываем некорректные даты рождения и корректные и вычисляем количество полных лет студентов
av_age['birthday'] = av_age['birthday'][av_age['birthday']>dt.date(1941, 1, 1)][av_age['birthday']<dt.date(2010, 1, 1)].apply((lambda x: (dt.date.today() - x).days/365.25))

# создаем новый датафрейм группируя его по названию курса и агреггируя колонку birthday для вычисления 
#'min', 'max', 'mean', 'median' нужных параметров 
av_age = pd.DataFrame(av_age.groupby('title')['birthday'].agg(['min', 'max', 'mean', 'median']))
av_age
```

### 2.1 Расчет прироста студентов на каждом курсе в каждом месяце от марта 2016 до июля 2019 года.

```python
## YOUR CODE HERE ###
# Сортируем датафрейм по выполненным домашним работам
merged_all_hw = merged_all[merged_all['is_homework'] == True].copy()

# Приводим столбец ['start_date'] к формату datetime
merged_all_hw['start_date'] = pd.to_datetime(merged_all_hw['start_date'], errors='coerce')
merged_all_hw['start_date'] = merged_all_hw['start_date'].dt.strftime('%Y/%m')

# Вычисляем минимальный модуль в котором лежит 1-я домашняя работа каждого курса
first_hw = merged_all_hw.groupby(['title']).agg({'module_number':'min'}).reset_index()
first_hw.columns = ['title', 'module_number']

# Соединяем два предыдущих датафрейма
sorted_merged_all_hw = merged_all_hw.merge(first_hw, left_on=['title', 'module_number'],
                                          right_on=['title', 'module_number'])


# Вычисляем прирост студентов
students_increase = sorted_merged_all_hw.groupby(['title','start_date']).agg({'student_id':'nunique'})\
                                      .sort_values('start_date')\
                                      .reset_index()

# Создаем сводную таблицу прироста студентов по каждому курсу
student_count_of_month = pd.pivot_table(students_increase, index=['start_date'], columns=['title'], 
                                        values = 'student_id', aggfunc=min, fill_value=0)

student_count_of_month
```

### 4.1 Расчет конверсии перехода студентов из одного модуля в другой на каждом курсе.

```python
# # ### YOUR CODE HERE ###

# Сортируем датафрей по выполненным домашним работам
conv_df = merged_all[merged_all['is_homework']==True].copy()

# Находим самую первую дату начала и самую последнюю дату конца домашней работы
conv_df = conv_df.groupby(['title', 'field', 'module_title', 'module_number', 'status', 'student_id'])\
                          .agg({'start_date':'min', 'finish_date':'max'}).reset_index()
# Оставляем в датафрейме строки в которых колонка status равна done 
model_done = conv_df[conv_df['status'] == 'done']
model_other = conv_df[conv_df['status'].isin(['done', 'start', 'fail', 'wait'])]

# Оставляем нужные нам колонки
model_done = model_done[['title', 'field', 'module_title', 'module_number', 'status', 'student_id']]
model_other = model_other[['title', 'field', 'module_title', 'module_number', 'status', 'student_id']]

# для датафрейма с колонкой status равный done и датафрейма со всеми остальными статусами вычисляем кол-во студентов
conversion_done = model_done.groupby(['title', 'field', 'module_number', 'module_title']).agg({'student_id':'nunique'})\
                                    .reset_index()
conversion_other = model_other.groupby(['title', 'field', 'module_number', 'module_title']).agg({'student_id':'nunique'})\
                                      .reset_index()

# Мерджим эти два датафрейма
conversion_all = conversion_done.merge(conversion_other, how='inner', on=['title', 'field', 'module_number','module_title'])

# Делаем сдвиг на 1 ячейку
conversion_all['student_id_x'] = conversion_all['student_id_x'].shift(1)

# Вычисляем конверсию
conversion_all['conversion'] = conversion_all['student_id_y']/conversion_all['student_id_x']
conversion_all['conversion'] = conversion_all['conversion'].fillna(1)
conversion_all.reset_index()
```
